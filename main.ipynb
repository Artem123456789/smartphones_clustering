```python
continious = ["battery_power", "clock_speed", "fc", "int_memory", "m_dep", "mobile_wt", "n_cores", "pc",
              "px_height", "px_width", "ram", "sc_h", "sc_w", "talk_time"]
discrete = df.drop(columns=continious).columns
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
continious_transformer = Pipeline(steps=[("scaler", StandardScaler())])
```

```python
preprocessor = ColumnTransformer(transformers=[("num", continious_transformer, continious)])
```

```python
inertias = []
clusters_range = np.arange(2,10)
for c in clusters_range:
    kmeans = KMeans(n_clusters=c, random_state=0)
    pipe = Pipeline(steps=[("preprocessor", preprocessor), ("kmeans", kmeans)])
    model = pipe.fit(df)
    inertias.append(kmeans.inertia_)
```

```python
plt.plot(clusters_range, inertias, marker="o")
```

```python
kmeans = KMeans(n_clusters=5, random_state=0)
pipe = Pipeline(steps=[("preprocessor", preprocessor), ("kmeans", kmeans)])
model = pipe.fit(df)
```

```python
results_df = pd.concat([df, pd.Series(kmeans.labels_, name="Labels")], axis=1)
```
